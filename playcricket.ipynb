```python
import pydotplus
from sklearn import tree
from IPython.display import Image, display

clf = tree.DecisionTreeClassifier() 
X = [[200, 1, 1, 0],
[200, 1, 1, 1],
[100, 1, 1, 0],
[300, 2, 1, 0],
[300, 3, 0, 0],
[300, 3, 0, 1],
[100, 3, 0, 1],
[200, 2, 1, 0],
[200, 3, 0, 0],
[300, 2, 0, 0],
[200, 2, 0, 1],
[100, 2, 1, 1],
[100, 1, 0, 0],
[300, 2, 1, 1]] 
Y = ['No', 'No', 'Yes', 'Yes', 'Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

clf = clf.fit(X, Y)
dot_data = tree.export_graphviz(clf, feature_names=['Outlook', 'Temperature', 'Humidity', 'Wind'], class_names=['No', 'Yes'], filled=True, rounded=True)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  
# Prediction
prediction = clf.predict([[300, 3,0,1]])
print('Prediction ', prediction) 

# Show graph
Image(graph.create_png())

#graph.write_pdf("playcricket.pdf")
#graph.write_png("playcricket.png")

```
